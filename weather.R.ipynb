```r
library("tidyverse")
library("here")
library("skimr")
library("janitor")
```

```r
weather_df <- read_csv("weather.csv")
```

```r
weather_df
```

```r
filter()
```

```r
#Average temperature in winter and in summer overall.
weather_df %>% 
summarize(winter_mean = (mean(JAN)+mean(FEB)+mean(MAR))/3, summer_mean = (mean(MAY)+mean(JUN)+mean(JUL))/3)
```

```r
#Compare the february and october temperatures using plots.
ggplot(data = weather_df)+
geom_point(aes(x=FEB, y=OCT))+
labs(title="FEB vs OCT")
```

```r
#Average temperature in winter and in summer in 1942.
weather_df %>% 
filter(YEAR == 1942) %>% 
summarize(winter_mean = (JAN+FEB+MAR)/3, summer_mean =(MAY+JUN+JUL)/3)
```

```r
#Temperature difference between winter and summer in 2001
weather_df %>% 
filter(YEAR == 2001) %>% 
summarize(diff_win_sum = abs(((JAN+FEB+MAR)/3)- ((MAY+JUN+JUL)/3)))
```

```r
#Find the avg temperature in 1971.
weather_df %>%
filter(YEAR==1971) %>% 
subset(select=c(JAN:DEC)) %>% 
rowMeans()
```

```r
#Find the avg temperature till now 
new_weather_df <- subset(weather_df, select = c(JAN:DEC))
mean(as.matrix(new_weather_df))
```

```r
#Which month has the lowest, highest temperatures, overall?
df <-subset(weather_df, select=c(JAN:DEC))
capture.output(cat("The lowest temperature recored in the month of ", names(df)[which.min(apply(df,MARGIN=2,min))]))
capture.output(cat("The highest temperature recorded in the month of ",names(df)[which.max(apply(df,MARGIN=2,max))]))
```
